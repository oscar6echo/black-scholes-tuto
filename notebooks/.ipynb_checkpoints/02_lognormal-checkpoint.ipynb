# Lognormal Distribution of Stock Prices

Under the risk-neutral measure $\mathbb{Q}$, the stock follows:
$$dS = (r-q)\,S\,dt + \sigma\,S\,dW^{\mathbb{Q}}$$
By Itô's lemma, $\ln(S_T/S_0) \sim \mathcal{N}\!\left((r-q-\tfrac{1}{2}\sigma^2)T,\;\sigma^2 T\right)$,
so $S_T$ is **lognormally distributed**.

This notebook:
1. Simulates GBM paths and plots the path fan
2. Compares the simulated terminal distribution to the theoretical lognormal PDF
3. Shows that log-returns $\ln(S_T/S_0)$ are normally distributed
4. Explores the effect of $\sigma$, $T$, and $q$ on the distribution

```python
import numpy as np
from scipy.stats import norm, lognorm
import plotly.graph_objects as go
from plotly.subplots import make_subplots

rng = np.random.default_rng(42)

# Base parameters
S0    = 100.0
r     = 0.05
q     = 0.02
sigma = 0.20
T     = 1.0
```

## 1. GBM Path Simulation

We use the **exact** discretisation (no Euler error):
$$S_{t+\Delta t} = S_t \cdot \exp\!\left((r-q-\tfrac{1}{2}\sigma^2)\,\Delta t + \sigma\,\sqrt{\Delta t}\,Z\right), \quad Z \sim \mathcal{N}(0,1)$$

The theoretical percentile bands use the exact lognormal distribution at each $t$.

```python
N_PATHS = 300
N_STEPS = 252   # daily
dt = T / N_STEPS
t  = np.linspace(0, T, N_STEPS + 1)

# Exact scheme: cumulative log returns
Z            = rng.standard_normal((N_PATHS, N_STEPS))
log_incr     = (r - q - 0.5*sigma**2)*dt + sigma*np.sqrt(dt)*Z
log_paths    = np.hstack([np.zeros((N_PATHS, 1)), np.cumsum(log_incr, axis=1)])
S_paths      = S0 * np.exp(log_paths)

fig = go.Figure()

# Faint individual paths
for i in range(min(80, N_PATHS)):
    fig.add_trace(go.Scatter(
        x=t, y=S_paths[i],
        mode='lines', line=dict(color='steelblue', width=0.6),
        opacity=0.25, showlegend=False
    ))

# Theoretical percentile bands
pctls  = [5, 25, 50, 75, 95]
styles = [
    dict(color='rgb(30,90,160)', width=0.8, dash='dash'),
    dict(color='rgb(30,90,160)', width=0.8, dash='dot'),
    dict(color='rgb(30,90,160)', width=2.0, dash='solid'),
    dict(color='rgb(30,90,160)', width=0.8, dash='dot'),
    dict(color='rgb(30,90,160)', width=0.8, dash='dash'),
]
mu_t  = np.log(S0) + (r - q - 0.5*sigma**2) * t
std_t = sigma * np.sqrt(np.where(t > 0, t, 1e-12))

for p, style in zip(pctls, styles):
    band = np.exp(norm.ppf(p/100)*std_t + mu_t)
    band[0] = S0
    fig.add_trace(go.Scatter(
        x=t, y=band, mode='lines',
        line=style, name=f'{p}th percentile'
    ))

fig.update_layout(
    title=f'GBM Paths under Q   (S\u2080={S0}, r={r:.0%}, q={q:.0%}, \u03c3={sigma:.0%}, T={T}y)',
    xaxis_title='Time (years)', yaxis_title='S\u209c',
    template='plotly_white', height=450,
    legend=dict(x=0.02, y=0.98)
)
fig.show()
```

## 2. Terminal Distribution vs Lognormal PDF

The theoretical distribution of $S_T$ is lognormal with parameters:
$$\mu_{\log} = \ln S_0 + (r-q-\tfrac{1}{2}\sigma^2)T, \qquad \sigma_{\log} = \sigma\sqrt{T}$$

The vertical dashed line marks the **forward price** $F = S_0\,e^{(r-q)T}$, which is
the mean of $S_T$ under $\mathbb{Q}$ (not the median — skewness of the lognormal).

```python
N_LARGE = 100_000
Z_large = rng.standard_normal(N_LARGE)
S_T     = S0 * np.exp((r - q - 0.5*sigma**2)*T + sigma*np.sqrt(T)*Z_large)

# Theoretical lognormal PDF
mu_log  = np.log(S0) + (r - q - 0.5*sigma**2)*T
sig_log = sigma * np.sqrt(T)
S_range = np.linspace(S_T.min()*0.4, np.percentile(S_T, 99.5), 600)
pdf_th  = lognorm.pdf(S_range, s=sig_log, scale=np.exp(mu_log))

forward = S0 * np.exp((r - q)*T)

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=S_T, histnorm='probability density',
    name='MC simulation', opacity=0.55,
    marker_color='steelblue', nbinsx=100
))
fig.add_trace(go.Scatter(
    x=S_range, y=pdf_th, mode='lines',
    name='Lognormal PDF', line=dict(color='firebrick', width=2.5)
))
fig.add_vline(
    x=forward, line_dash='dash', line_color='green', line_width=2,
    annotation_text=f'F = {forward:.1f}', annotation_position='top right'
)
fig.update_layout(
    title='Terminal Distribution of S_T',
    xaxis_title='S_T', yaxis_title='Probability Density',
    template='plotly_white', height=420
)
fig.show()

print(f'Simulated mean S_T:    {S_T.mean():.2f}   (theoretical: {forward:.2f})')
print(f'Simulated median S_T:  {np.median(S_T):.2f}   (theoretical: {np.exp(mu_log):.2f})')
```

## 3. Log-Returns are Normal

$S_T$ is lognormal; equivalently, $\ln(S_T/S_0)$ is **normal** with
mean $(r-q-\tfrac{1}{2}\sigma^2)T$ and std $\sigma\sqrt{T}$.

Note the **Itô correction** $-\tfrac{1}{2}\sigma^2$: the log-return mean is *less* than
the drift $(r-q)$ because the lognormal distribution is right-skewed.

```python
log_returns = np.log(S_T / S0)

mu_lr    = (r - q - 0.5*sigma**2)*T
sig_lr   = sigma*np.sqrt(T)
lr_range = np.linspace(log_returns.min(), np.percentile(log_returns, 99.5), 500)
lr_pdf   = norm.pdf(lr_range, loc=mu_lr, scale=sig_lr)

fig = make_subplots(1, 2,
    subplot_titles=['S_T  (lognormal)', 'ln(S_T / S\u2080)  (normal)'])

fig.add_trace(go.Histogram(x=S_T, histnorm='probability density',
    opacity=0.55, marker_color='steelblue', nbinsx=80, showlegend=False), row=1, col=1)
fig.add_trace(go.Scatter(x=S_range, y=pdf_th, mode='lines',
    line=dict(color='firebrick', width=2), showlegend=False), row=1, col=1)

fig.add_trace(go.Histogram(x=log_returns, histnorm='probability density',
    opacity=0.55, marker_color='coral', nbinsx=80, showlegend=False), row=1, col=2)
fig.add_trace(go.Scatter(x=lr_range, y=lr_pdf, mode='lines',
    line=dict(color='firebrick', width=2), showlegend=False), row=1, col=2)

fig.update_layout(template='plotly_white', height=400,
    title='Lognormal S_T vs Normal log-returns')
fig.show()

print(f'Log-return mean:  {log_returns.mean():.4f}  (theory: {mu_lr:.4f})')
print(f'Log-return std:   {log_returns.std():.4f}  (theory: {sig_lr:.4f})')
print(f'Ito correction:   -½σ²T = {-0.5*sigma**2*T:.4f}')
```

## 4. Effect of Volatility

Higher $\sigma$ widens the distribution symmetrically in log-space,
but because of the skewness of the lognormal, it pulls the right tail out further
than the left. The forward price $F$ (mean under $\mathbb{Q}$) does not change with $\sigma$,
but the median $e^{\mu_{\log}}$ shifts **left** (the Itô correction grows with $\sigma$).

```python
sigmas = [0.10, 0.20, 0.30, 0.45]
colors = ['royalblue', 'forestgreen', 'darkorange', 'crimson']
S_plot = np.linspace(20, 350, 800)

fig = go.Figure()
for sig, col in zip(sigmas, colors):
    mu_l  = np.log(S0) + (r - q - 0.5*sig**2)*T
    sig_l = sig * np.sqrt(T)
    pdf   = lognorm.pdf(S_plot, s=sig_l, scale=np.exp(mu_l))
    fig.add_trace(go.Scatter(
        x=S_plot, y=pdf, mode='lines',
        name=f'\u03c3 = {sig:.0%}', line=dict(color=col, width=2)
    ))

fig.add_vline(x=S0*np.exp((r-q)*T), line_dash='dash', line_color='black',
    annotation_text='F (unchanged)', annotation_position='top right')
fig.update_layout(
    title='Effect of Volatility on Terminal Distribution (T=1y)',
    xaxis_title='S_T', yaxis_title='Probability Density',
    template='plotly_white', height=420
)
fig.show()
```

## 5. Effect of Maturity

Longer maturity means more time for the Brownian motion to diffuse:
- Both $\mu_{\log}$ and $\sigma_{\log}^2 = \sigma^2 T$ grow with $T$
- The forward price $F = S_0 e^{(r-q)T}$ drifts upward
- The distribution spreads and shifts right

This is why longer-dated options are worth more (more uncertainty = more option value).

```python
maturities = [0.25, 0.5, 1.0, 2.0]
S_plot_T   = np.linspace(20, 450, 800)

fig = go.Figure()
for T_v, col in zip(maturities, colors):
    mu_l  = np.log(S0) + (r - q - 0.5*sigma**2)*T_v
    sig_l = sigma * np.sqrt(T_v)
    pdf   = lognorm.pdf(S_plot_T, s=sig_l, scale=np.exp(mu_l))
    F_v   = S0 * np.exp((r-q)*T_v)
    fig.add_trace(go.Scatter(
        x=S_plot_T, y=pdf, mode='lines',
        name=f'T = {T_v}y  (F={F_v:.0f})', line=dict(color=col, width=2)
    ))

fig.update_layout(
    title='Effect of Maturity on Terminal Distribution (\u03c3=20%)',
    xaxis_title='S_T', yaxis_title='Probability Density',
    template='plotly_white', height=420
)
fig.show()
```

## 6. Effect of Dividend Yield

The dividend yield $q$ reduces the forward growth rate: $F = S_0 e^{(r-q)T}$.
A higher $q$ shifts the distribution **left** without changing its shape (spread is unchanged).
This is why higher $q$ lowers call prices and raises put prices.

```python
div_yields = [0.00, 0.02, 0.05, 0.08]

fig = go.Figure()
for q_v, col in zip(div_yields, colors):
    mu_l  = np.log(S0) + (r - q_v - 0.5*sigma**2)*T
    sig_l = sigma * np.sqrt(T)
    pdf   = lognorm.pdf(S_plot, s=sig_l, scale=np.exp(mu_l))
    F_v   = S0 * np.exp((r-q_v)*T)
    fig.add_trace(go.Scatter(
        x=S_plot, y=pdf, mode='lines',
        name=f'q = {q_v:.0%}  (F={F_v:.0f})', line=dict(color=col, width=2)
    ))

fig.update_layout(
    title='Effect of Dividend Yield on Terminal Distribution (\u03c3=20%, T=1y)',
    xaxis_title='S_T', yaxis_title='Probability Density',
    template='plotly_white', height=420
)
fig.show()

print('Summary: q shifts the distribution left (lower forward) without changing its spread.')
print('sigma changes the spread; T changes both the spread and the centre.')
```
